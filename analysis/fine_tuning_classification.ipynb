```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
```

```python
dataframe = pd.read_csv("../logs/gpt2_classification_fine_tuned_20260126_011138.csv")
```

```python
dataframe.describe()
```

```python
def plot_losses(epochs_seen, examples_seen, train_losses, val_losses):
    fig, ax1 = plt.subplots(figsize=(5, 3))

    # Plot training and validation loss against epochs
    ax1.plot(epochs_seen, train_losses, label="Training loss")
    ax1.plot(epochs_seen, val_losses, linestyle="-.", label="Validation loss")
    ax1.set_xlabel("Epochs")
    ax1.set_ylabel("Loss")
    ax1.legend(loc="upper right")
    ax1.xaxis.set_major_locator(MaxNLocator(integer=True))  # only show integer labels on x-axis

    # Create a second x-axis for tokens seen
    ax2 = ax1.twiny()  # Create a second x-axis that shares the same y-axis
    ax2.plot(examples_seen, train_losses, alpha=0)  # Invisible plot for aligning ticks
    ax2.set_xlabel("Examples seen")

    fig.tight_layout()  # Adjust layout to make room
    
    plt.show()
```

```python
plot_losses(dataframe["Epochs"].values, dataframe["Examples Seen"], dataframe["Training Loss"], dataframe["Validation Loss"])
```

```python
def plot_accuracy(epochs_seen, examples_seen, train_accuracy, val_accuracy):
    fig, ax1 = plt.subplots(figsize=(5, 3))

    # Plot training and validation accuracy against epochs
    ax1.plot(epochs_seen, train_accuracy, label="Training accuracy")
    ax1.plot(epochs_seen, val_accuracy, linestyle="-.", label="Validation accuracy")
    ax1.set_xlabel("Epochs")
    ax1.set_ylabel("Loss")
    ax1.legend(loc="upper right")
    ax1.xaxis.set_major_locator(MaxNLocator(integer=True))  # only show integer labels on x-axis

    # Create a second x-axis for tokens seen
    ax2 = ax1.twiny()  # Create a second x-axis that shares the same y-axis
    ax2.plot(examples_seen, train_accuracy, alpha=0)  # Invisible plot for aligning ticks
    ax2.set_xlabel("Examples seen")

    fig.tight_layout()  # Adjust layout to make room
    
    plt.show()
```

```python
plot_accuracy(dataframe["Epochs"].values, dataframe["Examples Seen"], dataframe["Training Accuracy"], dataframe["Validation Accuracy"])
```
